```python
import pandas as pd
import re
from tabulate import tabulate

csv_file = 'data_with_details.csv'

df = pd.read_csv(csv_file)

# интересующие номера
desired_numbers = ['10', '12', '24', '31', '63']

def contains_desired_plate(detail_name):
    if "Тормозная пластина" not in detail_name:
        return False
    numbers = set(re.findall(r'№(\d+)', detail_name))
    return not set(desired_numbers).isdisjoint(numbers)

filtered_rows = df[df['Название детали'].apply(contains_desired_plate)]


selected_columns = ['Начало простоя', 'Длительность', 'Название детали']
filtered_and_selected = filtered_rows[selected_columns]

def extract_detail_number(name):
    match = re.search(r'№(\d+)', name)
    return match.group(1) if match else None

filtered_and_selected['Номер детали'] = filtered_and_selected['Название детали'].apply(extract_detail_number)

filtered_and_selected['Номер детали'] = pd.Categorical(filtered_and_selected['Номер детали'], categories=desired_numbers, ordered=True)

filtered_and_sorted = filtered_and_selected.sort_values(by='Номер детали')

print("Отфильтрованные и отсортированные строки:")
print(tabulate(filtered_and_sorted.drop(columns=['Номер детали']), headers='keys', tablefmt='rst', showindex=False))
```

## Изменение формата дат

```python
import pandas as pd

df = pd.read_csv('new_downtime_data.csv')

dates = df['Начало простоя'].tolist()


month_translation = {
    'янв': 'Jan', 'фев': 'Feb', 'мар': 'Mar', 'апр': 'Apr',
    'май': 'May', 'июн': 'Jun', 'июл': 'Jul', 'авг': 'Aug',
    'сен': 'Sep', 'окт': 'Oct', 'ноя': 'Nov', 'дек': 'Dec'
}

dates_translated = [date.replace(rus, eng) for date in dates for rus, eng in month_translation.items() if rus in date]

dates_datetime = pd.to_datetime(dates_translated, format='%d-%b-%y %H:%M:%S')

df['Начало простоя'] = dates_datetime

print(df.head())

df.to_csv('new_downtime_data.csv', index=False, encoding='utf-8')

print("Файл успешно сохранен как 'new_downtime_data.csv'")
```
